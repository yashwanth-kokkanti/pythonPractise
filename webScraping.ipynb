```
## Web Scraping Example using Beautiful Soup 
```

```
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```
url = 'https://www.goibibo.com/hotels/hotels-in-shimla-ct/'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}

## user agent is extracted from chrome console using - window.navigator.userAgent

response = requests.request('GET', url, headers=headers)

data = BeautifulSoup(response.text, 'html.parser')
print (data)
```

```
## Find all sections with Specified class name 

#cards_data = data.find_all('div', attrs={'class', 'width100 fl htlListSeo hotel-tile-srp-container hotel-tile-srp-container-template new-htl-design-tile-main-block'})

cards_data = data.find_all('div', attrs={'class' : 'SeoFiltersCardstyles__GoSafeCardBlockWrapper-sc-1wpwk3v-0 kKRaRc'})

print ('Total no of cards found : ', len(cards_data))

for card in cards_data:
  print (card)
```

```
for card in cards_data:
  text_1 = card.find('p')

  text_2 = card.find('a', attrs={'class': 'SeoFiltersCardstyles__CityLink-sc-1wpwk3v-5 jUUGt'})

  print (text_1.get_text(), " : ", text_2.get_text())
```

```
## Scrape Images 

images = data.find_all('img', src=True)

print ("Number of images : ", len(images))
```

```
for image in images:
  print (image)
```

```
## Fetching Images 

## Select src Tag
image_src = [x['src'] for x in images]

## select only .jpg format 

image_src = [x for x in image_src if x.endswith('png') or x.endswith('jpg')]

for image in image_src:
  print (image)
```
